## Match Data

```python
# get all match starts: we see that match 2 has 6 different match start, where 5 has duration less than 120 seconds. 
# In video they were invalid as well.
matches = p11_metadata[p11_metadata["type"] == "match"]
matches
```

```python
# So we remove match starts with duraction less than 10 minutes
matches = matches[matches["duration"]>600]
matches
```

## Highlight Data

```python
# highlight of match 2 is missing!!!!
highlights = metadata[(metadata["type"] == "highlight") & (metadata["perspective"]=="P11")]
highlights.head(10)
```

```python
highlights_duration = highlights[["match_id","duration"]].groupby(["match_id"]).sum()
highlights_duration["total_duration"] = matches["duration"].values[[0] + list(range(2,11))]
highlights_duration["ratio"] = highlights_duration["duration"]/ highlights_duration["total_duration"]
highlights_duration
```
